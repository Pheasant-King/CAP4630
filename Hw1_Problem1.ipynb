```
from mpl_toolkits import mplot3d
```

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```
def f(x, y):
  return (x ** 3 + y ** 3)
def f1(x, y):
  return (np.cos(x) + np.sin(y))
def f2(x, y):
  return (np.cos(x) ** 2 + np.sin(y))

x = np.linspace(-100, 100, 30)
y = np.linspace(-100, 100, 30)

X, Y = np.meshgrid(x, y)

Z = f(X, Y)
```

# Saddle Point at (0,0)

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

# Local Min at (2pi + pi, 2pi - pi / 2)


```
x = np.linspace(-100, 100, 30)
y = np.linspace(-100, 100, 30)

X, Y = np.meshgrid(x, y)

Z = f1(X, Y)
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

ax.view_init(20)
```

# Local Max at (2pi, 5pi/2)

```
x = np.linspace(-10, 10, 30)
y = np.linspace(-10, 10, 30)

X, Y = np.meshgrid(x, y)

Z = f2(X, Y)
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

ax.view_init(30,110)
```
